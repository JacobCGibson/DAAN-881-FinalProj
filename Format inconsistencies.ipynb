```python
import pandas as pd
import re
# Load all datasets to inspect their contents
territorial_changes_1816 = pd.read_csv("territorial_changes_since_1816_new.csv")
territorial_changes_1945 = pd.read_csv('territorial_changes_since_1945_new.csv')
hced_data = pd.read_csv('hced_data_new.csv')
battle_deaths = pd.read_csv('hced_data_new.csv')

# Show a summary of the datasets (first few rows and info)
datasets_summary = {
    "territorial_changes_1816_head": territorial_changes_1816.head(),
    "territorial_changes_1816_info": territorial_changes_1816.info(),
    "territorial_changes_1945_head": territorial_changes_1945.head(),
    "territorial_changes_1945_info": territorial_changes_1945.info(),
    "hced_data_head": hced_data.head(),
    "hced_data_info": hced_data.info(),
    "battle_deaths_head": battle_deaths.head(),
    "battle_deaths_info": battle_deaths.info()
}

datasets_summary
```

```python
# Check the unique values in the 'pop' and 'area' columns to look for any formatting issues
territorial_changes_1816['pop'] = pd.to_numeric(territorial_changes_1816['pop'], errors='coerce')
territorial_changes_1816['area'] = pd.to_numeric(territorial_changes_1816['area'], errors='coerce')

# Repeat the same for Territorial Changes 1945
territorial_changes_1945['pop'] = pd.to_numeric(territorial_changes_1945['pop'], errors='coerce')
territorial_changes_1945['area'] = pd.to_numeric(territorial_changes_1945['area'], errors='coerce')

# Check for any inconsistencies or NaNs generated due to conversion issues in 'pop' and 'area'
territorial_changes_1816 = territorial_changes_1816[territorial_changes_1816[['pop', 'area']].isna().any(axis=1)]
territorial_changes_1945 = territorial_changes_1945[territorial_changes_1945[['pop', 'area']].isna().any(axis=1)]



# Define a function to compute the average year for ranges like '1218-1219'
def clean_year(year):
    if isinstance(year, str):
        match = re.match(r'(\d+)-(\d+)', year)
        if match:
            # Take the average of the two years and round it
            start_year = int(match.group(1))
            end_year = int(match.group(2))
            return round((start_year + end_year) / 2)
    # If it's a valid year or not a range, return the year as is
    return year


hced_data['Year'] = hced_data['Year'].apply(clean_year)
battle_deaths['Year'] = battle_deaths['Year'].apply(clean_year)

# Convert the Year to numeric if necessary
hced_data['Year'] = pd.to_numeric(hced_data['Year'], errors='coerce')
battle_deaths['Year'] = pd.to_numeric(battle_deaths['Year'], errors='coerce')





```
